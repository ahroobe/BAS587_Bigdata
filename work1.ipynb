# 1. 코인 던지기 프로그램과 확률 분포

```python
%matplotlib nbagg
import random
from collections import defaultdict
from collections import OrderedDict
import matplotlib.pyplot as plt
```

```python
totalresult=defaultdict(lambda:0)
totalresult2=defaultdict(lambda:0)
for i in range(0,10000):
    result = {0:0, 1:0}
    for j in range(0,3000):
        rand = random.randint(1, 10)
        if rand%2==0:
            result[0]+=1
        else:
            result[1]+=1
    totalresult[result[0]*1.0/3000]+=1
    
for i in range(0,10000):
    result = {0:0, 1:0}
    for j in range(0,6000):
        rand = random.randint(1, 10)
        if rand%2==0:
            result[0]+=1
        else:
            result[1]+=1
    totalresult2[result[0]*1.0/6000]+=1
```

```python
sum(totalresult2.values())
```

```python
od = OrderedDict(sorted(totalresult.items()))
od2 = OrderedDict(sorted(totalresult2.items()))
```

```python
plt.plot(od.keys(),od.values(),color="r")
plt.plot(od2.keys(),od2.values(),color="b")
plt.show()
```

# 2. CAGR 구하기

\\( CAGR=(\frac{Endmonth}{Startmonth})^\frac{1}{Noyears}-1 \\)


```python
def getCAGR(end,start,noyear):
    CAGR = ((end*1.0/start)**(1.0/noyear))-1
    return CAGR
```

```python
getCAGR(2000,980,7)
```
